```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create a small sample dataset
data = {
    'State': ['Tamil Nadu', 'Maharashtra', 'Kerala', 'Delhi', 'Punjab'],
    'Date': ['2020-01', '2020-02', '2020-03', '2020-04', '2020-05'],
    'Frequency': ['Monthly'] * 5,
    'Estimated Unemployment Rate': [5.2, 6.1, 7.5, 8.2, 7.8],
    'Estimated Employed': [12000000, 18000000, 9000000, 7000000, 8000000],
    'Estimated Labour Participation Rate': [42.1, 40.3, 43.5, 39.8, 41.7],
    'Region': ['South', 'West', 'South', 'North', 'North'],
    'longitude': [78.6569, 75.7139, 76.2711, 77.1025, 75.3412],
    'latitude': [11.1271, 19.7515, 10.8505, 28.7041, 31.1471]
}

df = pd.DataFrame(data)

# Step 3: View the data
print("Sample dataset:")
print(df)

# Step 4: Plot unemployment trend by region
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Date', y='Estimated Unemployment Rate', hue='Region', marker='o')
plt.title("Sample: Unemployment Rate Over Time by Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 5: Show average unemployment by state
state_avg = df.groupby('State')['Estimated Unemployment Rate'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,6))
state_avg.plot(kind='barh', color='skyblue')
plt.title("Sample: Average Unemployment Rate by State")
plt.xlabel("Unemployment Rate (%)")
plt.tight_layout()
plt.show()
```
